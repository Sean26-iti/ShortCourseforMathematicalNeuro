### *Neurons*

- The neuron is the building block for the entire nervous system.
- Neurons are cells that can carry signals rapidly over distance. These signals are "action potentials" (APs, or spikes), relatively large fluctuations in the voltage across the cell membrane.
- APs propagate down the axon - an extension of the cell that forms a relatively long cable, compared to the soma (central cell body).
- Neurons exhibit significant variety in shape and size, but all have the same basic features: the soma, dendrites, which are the shorter and more numerously branching extensions that receive signals from other neurons, and the axon, which carries signals to other neurons, and may also be branched.
- Principal neurons are often excitatory.
- Cortical pyramidal neurons, examples of projection neurons, are the primary excitatory neurons of the cerebral cortex. (Pyramidal because of their shape, primary workhorse of the nervous system)
- Interneurons have shorter axons and usually provide inhibitory inputs.

#### *Action Potentials*

- Large number of ion channels in neuron membrane.
- Most are selective for a single ion, e.g. $Ca^{2+}$, $Na^{+}, K^{+}$, $Cl^{-}$
- These pores control the flow of each ion by opening and closing in response both to the voltage across the membrane and to internal or external signals.
- The membrane voltage - the difference between the electrical potential inside and outside the cell - is the key variable.
- A typical resting potential (the potential inside the neuron relative to the extracellular fluid measured when the neuron is not spiking) is approximately 70 mV.
- An AP, or a spike, is a fluctuation on the order of 100 Mv that lasts about a millisecond.
- The resting potential is maintained by ion pumps, which remove *$Na^{+}$* from the cell and bring *$K^{+}$* into the cell.
- Ions flow along concentration gradients, high to low, and along voltage gradients, with negative ions moving towards higher potential, and the inverse.
- A positive current is defined outward, and consists of positive ions flowing out or negative ions flowing in.
- A positive current *hyperpolarises* the cell (makes the voltage more negative), and a negative inward current *depolarises* it (makes the voltage more positive).
- Depolarisation leads to spikes.
- The spike is typically caused by *$Na^{+}$* channels opening quickly, causing an inrush of the ions, which raises the voltage and creates the spike, followed by *$K^{+}$* channels opening, allowing an outflow of those ions, which drops the membrane voltage below resting potential.
- After a spike, there is an *absolute refractory period*, during which the cell cannot spike, followed by a *relative refractory period* as ions concentrations recover, during which it is very difficult, but not impossible, to spike.
- Small changes in voltage insufficient to trigger a spike are called *subthreshold potentials*.
- APs regenerate along the axon and travel long distances without attenuation.
- Wave propagation in this active medium is quite different from propagation through a passive medium like that of sound waves in still air.
